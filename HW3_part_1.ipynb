```python
def exp(x) -> int:
  if x != x:
    return x

  else:
    count_month = 0

    if x[1].startswith("л") or x[1].startswith("г"):
      count_month += int(x[0]) * 12

    elif x[1].startswith("м"):
      count_month += int(x[0])

    try:
      int(x[2])
    except ValueError:
      count_month += 0
    else:
      count_month += int(x[2])

    return count_month
```

```python
df_hh['Опыт работы в мес.'] = df_hh['Опыт работы в мес.'].apply(exp)
```
